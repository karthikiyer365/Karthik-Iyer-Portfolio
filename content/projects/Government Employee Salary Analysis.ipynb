# Storyline

## Context

Government pay data is public but scattered behind APIs and rarely examined as a time series; raw salary tables obscure how compensation actually shifts across departments, roles, and years.

## Workflow

Live API ingestion & pagination → cleaning and feature engineering → multi-year exploratory analysis → an interactive department dashboard.

## Technical Implementation

- **Ingestion:** pulled Louisville Metro salary records (2015–2023) directly from the city's open ArcGIS REST API, handling year filters and offset pagination with `requests` + `pandas.json_normalize`.
- **Cleaning & features:** column normalization, null handling, per-employee IDs via `groupby`, and a derived `Calculated_Total` (regular + incentive + overtime).
- **EDA:** per-employee salary trends, year-over-year mean shifts, department distribution pies, correlation heatmaps, and rate-vs-total scatter plots with Matplotlib, Seaborn, and Plotly.
- **Dashboard:** a Plotly Dash app with a department selector plus a build-your-own-chart panel (choose X/Y axes and histogram vs. scatter) over label-encoded categorical fields.

## Outcomes

Turned a raw public-sector salary API into a reproducible pipeline and an interactive dashboard, surfacing department-level pay structure and multi-year compensation trends for Louisville Metro.
