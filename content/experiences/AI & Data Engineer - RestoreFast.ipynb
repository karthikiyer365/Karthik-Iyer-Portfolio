# Storyline

## Context

RestoreFast's operations ran across disconnected third-party tools with no unified data layer — worker discovery, compliance, and client communication each lived in a silo, slowing project execution and burying staff in manual triage.

## Workflow

The platform connects worker discovery → evaluation & matching → onboarding & compliance → live project communication, with an AI knowledge layer and a 24/7 hotline agent across the flow to triage inbound work and surface answers from internal data.

## Technical Implementation

- **Agentic matching (DS-CAR):** multi-step agentic flows that identify, evaluate, and select worker profiles against project requirements across five dimensions, wired into integrated communication channels.
- **Integration layer:** 5+ third-party platforms (Gmail, Drive, Nutshell, Xero, Twilio) unified through custom MCP tools, RAG connectors, and REST services.
- **Knowledge platform:** ingestion and retrieval layer over 3 connected Agentic/RAG/MCP setups on Pinecone with custom schema design.
- **Hotline agent:** 24/7 LLM-backed tier-1 call triage on a queued audio-processing backend with Twilio, cutting routine escalations ~60%.
- **Communications layer:** event routing from 15+ internal tools to SMS, Voice, and Email through a containerized Twilio integration.
- **Compliance hub:** document hub with expiration tracking and attachment references across 5+ active projects on Supabase + S3.

## Outcomes

Consolidated five fragmented operational areas into connected systems, reduced routine call escalations ~60%, and gave the team a single AI-assisted layer across worker discovery, compliance, and communication.
