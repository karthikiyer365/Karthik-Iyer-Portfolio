```python
# restorefast_synthetic_charts.py
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

# -----------------------------
# Radar chart: Role readiness
# -----------------------------
def radar_role_readiness():
    """
    Radar plot comparing role-readiness across key dimensions.
    Dimensions informed by your work: LLM Ops, ETL/DataEng, Infra/Systems, ML Modeling,
    Comms/Integration, Deployment/CI, Product Sense.
    Scores are synthetic but anchored to documented strengths:
    - Strong on Data Engineering, Agentic/LLM Ops, Comms, Deployment.
    """
    categories = ["LLM Ops", "ETL / DataEng", "Systems / Infra", "ML Modeling", "Comms / Integration", "CI/CD & Deploy", "Product Sense"]
    # Synthetic scores 0-10 (anchored to resume/notes)
    ai_engineer = [9, 7, 6, 8, 8, 7, 6]         # strong LLM & ML
    data_engineer = [5, 10, 9, 6, 6, 9, 5]     # strong ETL, infra, CI/CD
    systems_analyst = [6, 8, 10, 5, 7, 8, 7]   # operations / infra leaning

    fig = go.Figure()
    fig.add_trace(go.Scatterpolar(r=ai_engineer, theta=categories, fill='toself', name='AI Engineer'))
    fig.add_trace(go.Scatterpolar(r=data_engineer, theta=categories, fill='toself', name='Data Engineer'))
    fig.add_trace(go.Scatterpolar(r=systems_analyst, theta=categories, fill='toself', name='Systems Analyst'))
    fig.update_layout(
        polar=dict(radialaxis=dict(visible=True, range=[0,10])),
        title="Role Readiness Radar (synthetic) — RestoreFast",
        height=550
    )
    fig.show()


# -----------------------------
# Heatmap: Project × Technology intensity
# -----------------------------
def heatmap_project_tech():
    """
    Heatmap showing intensity (0-10) of tech usage per project/module.
    Rows = RestoreFast modules (RF-AR/Agentic, Voice Agent, Skillbridge, ETL Pipelines, PostHog/Instrumentation)
    Columns = techs (RAG/MCP, Agentic Dev, Data Eng, ML, Comms/Twilio, Docker/CI, Trigger.dev, OpenRouter)
    Values are synthetic but follow the documented responsibilities and tools in your docs.
    """
    projects = ["RF-AR / Agentic Brain", "Voice Agent (Ava Brooks)", "Skillbridge (Matching & Vetting)", "ETL Pipelines & Data Accuracy", "PostHog & Observability"]
    techs = ["RAG/MCP", "Agentic Dev", "Data Engineering", "ML Modeling", "Comms (Twilio/SendGrid)", "Docker / CI", "Trigger.dev", "OpenRouter"]
    # Synthetic intensity matrix (rows correspond to projects)
    mat = np.array([
        # RAG   Agentic  DataEng ML   Comms  Docker CI  OpenRouter
        [9,     10,      6,      7,    4,     6,    9,   10],   # RF-AR
        [2,     8,       4,      5,    10,    7,    6,    6],    # Voice Agent
        [3,     7,       9,      6,    8,     6,    5,    4],    # Skillbridge
        [1,     3,       10,     4,    2,     10,   3,    2],    # ETL Pipelines
        [1,     2,       4,      2,    6,     3,    2,    2]     # PostHog
    ])

    fig = go.Figure(data=go.Heatmap(
        z=mat,
        x=techs,
        y=projects,
        colorscale='Viridis',
        zmin=0, zmax=10
    ))
    fig.update_layout(title="Project × Technology Intensity (synthetic) — RestoreFast",
                      xaxis_title="Technology / Stack",
                      yaxis_title="Project / Module",
                      height=480)
    fig.show()


# -----------------------------
# Stacked bar: monthly comms by channel + conversion to intake
# -----------------------------
def stacked_comms_monthly():
    """
    Stacked bar chart of communication volumes by channel (Email, SMS, Voice, Slack, Push)
    across six months. Also shows a line for monthly intake conversions (synthetic).
    Anchor: Voice has higher conversion (intakes often come from calls), and voice agent automation
    increases over time as the hotline is productionized.
    """
    months = ["Jul 2025", "Aug 2025", "Sep 2025", "Oct 2025", "Nov 2025", "Dec 2025"]

    # Synthetic channel counts (these reflect the introduction and ramp of voice agent + comms package)
    email  = np.array([220, 210, 200, 195, 180, 160])
    sms    = np.array([120, 130, 140, 150, 160, 170])
    voice  = np.array([40, 60, 120, 180, 240, 300])   # voice grows as agent productionizes (Ava Brooks)
    slack  = np.array([12, 15, 18, 20, 22, 25])
    push   = np.array([30, 35, 40, 45, 50, 60])

    # Synthetic intake conversion (counts) per month derived from weighted channel conversion
    # Voice has higher per-touch conversion, others lower.
    conversion = np.round(0.05*email + 0.07*sms + 0.22*voice + 0.01*slack + 0.03*push).astype(int)

    df = pd.DataFrame({
        "Month": months,
        "Email": email,
        "SMS": sms,
        "Voice": voice,
        "Slack": slack,
        "Push": push,
        "Conversions": conversion
    })

    # stacked bar
    fig = go.Figure()
    fig.add_trace(go.Bar(x=df['Month'], y=df['Email'], name='Email', marker_color='#636EFA'))
    fig.add_trace(go.Bar(x=df['Month'], y=df['SMS'], name='SMS', marker_color='#EF553B'))
    fig.add_trace(go.Bar(x=df['Month'], y=df['Voice'], name='Voice', marker_color='#00CC96'))
    fig.add_trace(go.Bar(x=df['Month'], y=df['Slack'], name='Slack', marker_color='#AB63FA'))
    fig.add_trace(go.Bar(x=df['Month'], y=df['Push'], name='Push', marker_color='#FFA15A'))

    # overlay conversion line (secondary y-axis)
    fig.add_trace(go.Scatter(x=df['Month'], y=df['Conversions'], name='Estimated Intakes', mode='lines+markers',
                             marker=dict(size=10, color='black'), yaxis='y2'))

    # layout with second axis
    fig.update_layout(
        title="Monthly Communication Volume by Channel (synthetic) — RestoreFast\n(Voice agent ramp raises volume & conversion)",
        barmode='stack',
        xaxis_title="Month",
        yaxis=dict(title='Message Count'),
        yaxis2=dict(title='Estimated Intakes (count)', overlaying='y', side='right', range=[0, max(df['Conversions']) * 1.6]),
        legend=dict(x=0.01, y=0.99),
        height=520
    )
    fig.show()


# -----------------------------
# Run the charts
# -----------------------------
if __name__ == "__main__":
    radar_role_readiness()
    heatmap_project_tech()
    stacked_comms_monthly()

def exp_chart():
    import plotly.graph_objects as go
    from plotly.subplots import make_subplots
    import numpy as np

    np.random.seed(42)

    # -----------------------------
    # Data Definitions
    # -----------------------------
    industries = ["IT/Systems support", "Finance", "Operations", "Sports"]
    industry_values = [3, 7, 9, 2]  # relative counts (synthetic)

    tech_stacks = ["RAG/MCP", "Agentic Dev", "Data Engineering", "ML-Reg", "ML-Clf", "Data Analysis", "Market Research", "Software Dev"]
    tech_usage = [7, 6, 9, 8, 5, 6, 4, 6]  # relative usage scores

    heatmap_data = np.array([
        [2, 0, 6, 4, 2, 1, 0, 0],   # IT/Systems
        [1, 1, 3, 6, 2, 9, 4, 2],   # Finance
        [9, 7, 10, 4, 1, 2, 3, 10], # Operations (dominant)
        [0, 0, 7, 10, 8, 9, 4, 8]   # Sports
    ])

    career_progression = [1, 2, 3, 4, 5]
    career_labels = [
        "Financial Analyst Intern",
        "Teaching Assistant",
        "Technical Systems Assistant",
        "Performance Analyst",
        "Data Systems Engineer"
    ]

    # -----------------------------
    # Create Subplots Layout
    # -----------------------------
    fig = make_subplots(
        rows=2, cols=2,
        specs=[[{"type": "domain"}, {"type": "xy"}],
               [{"type": "heatmap"}, {"type": "xy"}]],
        subplot_titles=(
            "Industry Distribution",
            "Tech Stack Usage",
            "Industry × Tech Heatmap",
            "Career Progression Timeline"
        )
    )

    # Pie Chart
    fig.add_trace(
        go.Pie(labels=industries, values=industry_values, hole=0.35, sort=False),
        row=1, col=1
    )

    # Bar Chart
    fig.add_trace(
        go.Bar(x=tech_stacks, y=tech_usage, marker_color='teal'),
        row=1, col=2
    )

    # Heatmap
    fig.add_trace(
        go.Heatmap(
            z=heatmap_data,
            x=tech_stacks,
            y=industries,
            colorscale='Viridis'
        ),
        row=2, col=1
    )

    # Line Chart
    fig.add_trace(
        go.Scatter(
            x=career_labels,
            y=career_progression,
            mode="lines+markers",
            line=dict(color='darkorange')
        ),
        row=2, col=2
    )

    fig.update_layout(
        height=850,
        title_text="Experience Summary Analysis Dashboard (synthetic)",
        showlegend=False,
        margin=dict(t=100)
    )
    fig.show()
```
