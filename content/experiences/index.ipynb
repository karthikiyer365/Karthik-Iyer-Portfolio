```python
def exp_chart():
  import plotly.graph_objects as go
  from plotly.subplots import make_subplots
  import numpy as np

  # -----------------------------
  # Data Definitions
  # -----------------------------
  industries = ["IT/Systems support", "Finance", "Operations", "Sports"]
  industry_values = [3, 1, 5, 2]

  tech_stacks = ["RAG/MCP", "Agentic Development", "Data Engineer", "ML-Regression", "ML-Classification", "Data Analysis", "Market/Product Research", "Software Development"]
  tech_usage = [7, 6, 9, 8, 5, 6, 4, 6]

  heatmap_data = np.array([
      [2,0,9,4,2,1,0,0],  # EduTech
      [0,0,2,6,2,9,4,2],  # Finance
      [9,7,8,4,0,0,2,10],  # Ops
      [0,0,7,10,8,9,4,8]   # Sports
  ])

  career_progression = [1,2,3,4,5]
  career_labels = [
      "Financial Analyst Intern",
      "Big Data Interpretation - Teaching Assistant",
      "Technical Systems Assistant",
      "Sales & Marketing Operations Lead",
      "AI & Data Engineer"
  ]

  # -----------------------------
  # Create Subplots Layout
  # -----------------------------
  fig = make_subplots(
      rows=2, cols=2,
      specs=[[{"type": "domain"}, {"type": "xy"}],
            [{"type": "heatmap"}, {"type": "xy"}]],
      subplot_titles=(
          "Industry Distribution",
          "Tech Stack Usage",
          "Industry Tech Overlap",
          "Career Progression Timeline"
      )
  )

  # Pie Chart
  fig.add_trace(
      go.Pie(labels=industries, values=industry_values, hole=0.35),
      row=1, col=1
  )

  # Bar Chart
  fig.add_trace(
      go.Bar(x=tech_stacks, y=tech_usage),
      row=1, col=2
  )

  # Heatmap
  fig.add_trace(
      go.Heatmap(
          z=heatmap_data,
          x=tech_stacks,
          y=industries
      ),
      row=2, col=1
  )

  # Line Chart
  fig.add_trace(
      go.Scatter(
          x=career_labels,
          y=career_progression,
          mode="lines+markers"
      ),
      row=2, col=2
  )

  fig.update_layout(
      height=800,
      title_text="Experience Summary Analysis Dashboard",
      showlegend=False
  )

  fig.show()

def summary_chart():
    """
    summary_chart()
    Creates a two-panel Plotly chart (left: radar plot, right: KDE plot) summarizing
    Experiences (emerald green), Projects (deep shiny blue), and Volunteer (gold).
    Background is transparent. Font is Helvetica, color white, font size ~4.

    The function builds synthetic KDE distributions centered on the skill-level
    scores provided and displays the radar and KDE side-by-side. The legend is
    shown only for the KDE plot (right subplot). Returns the Plotly figure.
    """
    # --- imports inside function as requested ---
    import numpy as np
    import plotly.graph_objects as go
    from plotly.subplots import make_subplots
    from scipy.stats import gaussian_kde

    # -----------------------------
    # Color defs
    # -----------------------------
    color_experience = "#50C878"  # Emerald green
    color_projects = "#0057B7"    # Deep shiny blue
    color_volunteer = "#FFD700"   # Gold yellow

    def hex_to_rgba(hex_color, alpha=1.0):
        """Convert hex string to rgba() string for plotly fillcolors."""
        hex_color = hex_color.lstrip("#")
        r, g, b = tuple(int(hex_color[i:i+2], 16) for i in (0, 2, 4))
        return f"rgba({r},{g},{b},{alpha})"

    # -----------------------------
    # Radar data (exact user-provided indices)
    # Order: projects, experiences, volunteer
    # -----------------------------
    categories = [
        "Data Analysis",
        "Machine Learning",
        "Data Engineering",
        "AI Integration",
        "Schema Design",
        "Front end Development",
        "Backend Development",
        "Project Management",
        "Team Leadership"
    ]

    projects_vals = [90, 80, 50, 40, 70, 10, 30, 40, 30]
    experiences_vals = [65, 50, 87.5, 90, 70, 50, 55, 60, 50]
    volunteer_vals = [30, 0, 10, 0, 0, 30, 40, 80, 100]

    # For closed radar polygons: repeat first element at end
    cat_closed = categories + [categories[0]]
    proj_closed = projects_vals + [projects_vals[0]]
    exp_closed = experiences_vals + [experiences_vals[0]]
    vol_closed = volunteer_vals + [volunteer_vals[0]]

    # -----------------------------
    # Build synthetic KDE data
    # We'll create per-category samples by making samples around each skill score
    # then concatenate them into one array per category (projects/experiences/volunteer).
    # -----------------------------
    rng = np.random.default_rng(42)

    def build_samples(scores, samples_per_skill=40, sigma=6.0):
        """Create synthetic sample points around each skill score."""
        all_samples = []
        for s in scores:
            # If skill score is exactly 0, produce a small cluster near 0 to avoid empty KDE
            if s == 0:
                samp = rng.normal(loc=1.5, scale=sigma * 0.25, size=samples_per_skill)
            else:
                samp = rng.normal(loc=s, scale=sigma, size=samples_per_skill)
            # Clip to 0-100
            samp = np.clip(samp, 0, 100)
            all_samples.append(samp)
        return np.concatenate(all_samples)

    samples_projects = build_samples(projects_vals, samples_per_skill=40, sigma=6.0)
    samples_experiences = build_samples(experiences_vals, samples_per_skill=40, sigma=6.0)
    samples_volunteer = build_samples(volunteer_vals, samples_per_skill=40, sigma=6.0)

    # KDE estimation grid
    x_grid = np.linspace(0, 100, 400)

    def kde_density(samples, xgrid):
        kde = gaussian_kde(samples, bw_method='scott')
        dens = kde.evaluate(xgrid)
        # normalize so max density is consistent across series (for visual balance)
        dens = dens / dens.max() if dens.max() > 0 else dens
        return dens

    dens_proj = kde_density(samples_projects, x_grid)
    dens_exp = kde_density(samples_experiences, x_grid)
    dens_vol = kde_density(samples_volunteer, x_grid)

    # -----------------------------
    # Create subplots (polar left, xy right)
    # -----------------------------
    fig = make_subplots(
        rows=1, cols=2,
        specs=[[{"type": "polar"}, {"type": "xy"}]],
        column_widths=[0.45, 0.55],
        horizontal_spacing=0.08,
        subplot_titles=("Skill Radar (summary)", "Skill Distribution (KDE)")
    )

    # -----------------------------
    # Radar traces (no legend display)
    # -----------------------------
    fig.add_trace(
        go.Scatterpolar(
            r=exp_closed,
            theta=cat_closed,
            name="Experiences",
            mode="lines+markers",
            line=dict(color=color_experience, width=2),
            marker=dict(symbol="circle", size=5, color=color_experience),
            fill="toself",
            fillcolor=hex_to_rgba(color_experience, 0.18),
            showlegend=False,
            hovertemplate="%{theta}: %{r}<extra>Experiences</extra>"
        ),
        row=1, col=1
    )

    fig.add_trace(
        go.Scatterpolar(
            r=proj_closed,
            theta=cat_closed,
            name="Projects",
            mode="lines+markers",
            line=dict(color=color_projects, width=2),
            marker=dict(symbol="square", size=5, color=color_projects),
            fill="toself",
            fillcolor=hex_to_rgba(color_projects, 0.18),
            showlegend=False,
            hovertemplate="%{theta}: %{r}<extra>Projects</extra>"
        ),
        row=1, col=1
    )

    fig.add_trace(
        go.Scatterpolar(
            r=vol_closed,
            theta=cat_closed,
            name="Volunteer",
            mode="lines+markers",
            line=dict(color=color_volunteer, width=2),
            marker=dict(symbol="diamond", size=5, color=color_volunteer),
            fill="toself",
            fillcolor=hex_to_rgba(color_volunteer, 0.18),
            showlegend=False,
            hovertemplate="%{theta}: %{r}<extra>Volunteer</extra>"
        ),
        row=1, col=1
    )

    # Polar layout style
    fig.update_polars(
        bgcolor='rgba(0,0,0,0)',  # transparent
        radialaxis=dict(range=[0, 100], tickvals=[0,20,40,60,80,100],
                        tickfont=dict(color='white', family='Helvetica', size=4),
                        gridcolor='rgba(255,255,255,0.06)',
                        gridwidth=0.8,
                        linecolor='rgba(255,255,255,0.06)'),
        angularaxis=dict(tickfont=dict(color='white', family='Helvetica', size=4),
                         rotation=90, direction="clockwise")
    )

    # -----------------------------
    # KDE traces on right subplot (legend displayed only here)
    # Use same colours and include fill
    # -----------------------------
    # Projects KDE
    fig.add_trace(
        go.Scatter(
            x=x_grid, y=dens_proj,
            mode='lines',
            name='Projects',
            line=dict(color=color_projects, width=2),
            fill='tozeroy',
            fillcolor=hex_to_rgba(color_projects, 0.18),
            hovertemplate='Score: %{x:.1f}<br>Density: %{y:.3f}<extra>Projects</extra>'
        ),
        row=1, col=2
    )

    # Experiences KDE
    fig.add_trace(
        go.Scatter(
            x=x_grid, y=dens_exp,
            mode='lines',
            name='Experiences',
            line=dict(color=color_experience, width=2),
            fill='tozeroy',
            fillcolor=hex_to_rgba(color_experience, 0.18),
            hovertemplate='Score: %{x:.1f}<br>Density: %{y:.3f}<extra>Experiences</extra>'
        ),
        row=1, col=2
    )

    # Volunteer KDE
    fig.add_trace(
        go.Scatter(
            x=x_grid, y=dens_vol,
            mode='lines',
            name='Volunteer',
            line=dict(color=color_volunteer, width=2),
            fill='tozeroy',
            fillcolor=hex_to_rgba(color_volunteer, 0.18),
            hovertemplate='Score: %{x:.1f}<br>Density: %{y:.3f}<extra>Volunteer</extra>'
        ),
        row=1, col=2
    )

    # -----------------------------
    # Layout & styling (transparent background, white Helvetica fonts size 4)
    # -----------------------------
    fig.update_layout(
        paper_bgcolor='rgba(0,0,0,0)',
        plot_bgcolor='rgba(0,0,0,0)',
        font=dict(family='Helvetica', color='white', size=4),
        margin=dict(l=20, r=20, t=60, b=20),
        title=dict(text="RestoreFast — Skills Summary (synthetic)", x=0.5, xanchor='center', font=dict(family='Helvetica', color='white', size=8))
    )

    # Right subplot (KDE) axis styling
    fig.update_xaxes(title_text="Skill Score (0-100)", row=1, col=2,
                     tickfont=dict(color='white', family='Helvetica', size=4),
                     title_font=dict(color='white', family='Helvetica', size=4),
                     showgrid=False)
    fig.update_yaxes(title_text="Normalized Density", row=1, col=2,
                     tickfont=dict(color='white', family='Helvetica', size=4),
                     title_font=dict(color='white', family='Helvetica', size=4),
                     showgrid=False)

    # Legend only for KDE plot (we will keep the legend visible but style it small)
    fig.update_layout(
        legend=dict(
            orientation="v",
            yanchor="top",
            y=0.95,
            xanchor="left",
            x=1.02,
            font=dict(family='Helvetica', color='white', size=7),  # legend font slightly larger to read
            bgcolor='rgba(0,0,0,0)'
        )
    )

    # Ensure hover labels have white text
    fig.update_traces(hoverlabel=dict(font=dict(family='Helvetica', color='white', size=6)),
                      selector=dict(type='scatter'))


    # Return the figure; caller can show or save it
    return fig

# If run as a script, display immediately (uncomment to show)
# fig = summary_chart()
# fig.show()
```
